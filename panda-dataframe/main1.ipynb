### การเลือกข้อมูลใน DataFrame

#### การเลือกข้อมูลโดยระบุแถวและคอลัมน์
- หากไม่ได้ระบุชื่อคอลัมน์ ให้อ้างอิงด้วยลำดับคอลัมน์
- **pd[ลำดับคอลัมน์][ลำดับแถว]**
- สังเกตุว่าจะอ้างอิงด้วย **ลำดับคอลัมน์** ก่อน

```python
b8 = [
    [10,20,30],
    [40,50,60],
    [70,80,90]
]

b9 = pd.DataFrame(b8,index=list("abc"))

display(b9)
display(b9[0] , b9[1] , b9[2])

print(b9[0][0])
print(b9[1][1])
```

- ถ้าหากระบุชื่อคอลัมน์ สามารถอ้างถึงข้อมูลได้แบบนี้ **pd[ชื่อคอลัมน์][ลำดับแถว]** หรือ **pd.ชื่อคอลัมน์[ลำดับแถว]**
- **ถ้าหากกำหนดชื่อคอลัมน์เป็น 0 1 2 3.... จะ error**

```python
b10 = {
    "one":np.array([1,3,5,7]),
    "two":np.array([2,4,6,8]),
    "three":np.array([1,4,7,10])
}

c1 = pd.DataFrame(b10,index=list("abcd"))

display(c1)
display(c1["one"] , c1.two)

print(c1["one"]["a"])
print(c1["three"][3])
```

#### การเลือกข้อมูลโดยใช้โอเปอเรเตอร์ at และ iat
- **iat** ใช้เมื่อระบุแถวและคอลัมน์เป้าหมายด้วยชื่ออ้างอิง  **pd.iat[ลำดับแถว,ลำดับคอลัมน์]**
- **at** ใช้เมื่อระบุแถวและคอลัมน์เป้าหมายด้วยเลขลำดับ **pd.at[แถว,คอลัมน์]**

```python
c2 = {
    "one":np.array([1,3,5,7]),
    "two":np.array([2,4,6,8]),
    "three":np.array([1,4,7,10])
}

c3 = pd.DataFrame(b10,index=("a","b","c","d"))

display(c3)
print(c3.at["a","three"])
print(c3.at["b","two"])
print(c3.iat[3,2])
print(c3.iat[1,1])
```

#### การเลือกข้อมูลโดยวิธี index slicing
- ใช้กำหนดช่วงลำดับแถว วิธีนี้กำหนดได้เลือกเพียงช่วงแถวเท่านั้น **pd[start:stop:step]**

```python
c4 = {
    "one":np.random.randint(1,30,6),
    "two":np.random.randint(1,30,6),
    "three":np.random.randint(1,30,6),
    "four":np.random.randint(1,30,6)
}

c5 = pd.DataFrame(c4,index=list("abcdeg"))
display(c5 , c5[0:3] , c5[0::2] , c5["a":"b"])
```

#### การเลือกข้อมูลโดยวิธี fancy index + index slicing
- **pd[["col1","col2","col3"]]** เลือกเฉพาะคอลัมน์ของทุกแถว
- **pd[["col1","col2"]][start:stop:step]** หรือ **pd[start:stop:step][["col1","col2"]]**  เลือกเฉพาะคอลัมน์ ในช่วงลำดับแถวที่ต้องการ

```python
c6 = {
    "one":np.random.randint(1,30,6),
    "two":np.random.randint(1,30,6),
    "three":np.random.randint(1,30,6),
    "four":np.random.randint(1,30,6),
    "five":np.random.randint(1,30,6),
    "six":np.random.randint(1,30,6),
}

c7 = pd.DataFrame(c6,index=list("abcdef"))
display(c7 , c7[["one","three","five"]] , c7[["one","two"]][0::2] , c7[2:5][["five","six"]] , c7["a":"c"][["three","four"]])
```

#### การใช้ loc และ iloc ร่วมกับ fancy index
- หากต้องการเลือกเฉพาะบางแถวของทุกคอลัมน์ โดยระบุค่าแบบ label index ก็ใช้ loc **pd.loc[["row1","row2","row3"]]**
- หากต้องการเลือกเฉพาะบางแถวของทุกคอลัมน์ โดยระบุค่าแบบ position index ก็ใช้ loc **pd.iloc[[row1,row2,row3]]**

```python
c8 = {
    "one":np.random.randint(1,30,6),
    "two":np.random.randint(1,30,6),
    "three":np.random.randint(1,30,6),
    "four":np.random.randint(1,30,6),
    "five":np.random.randint(1,30,6),
}

c9 = pd.DataFrame(c8,index=list("abcdef"))
display(c9 , c9.loc[["a","c","e"]] , c9.iloc[[0,1,2]])
```

- การใช้ loc , iloc ต้องระบุแถวก่อนคอลัมน์เสมอ
- หากต้องการเลือกบางแถวและบางคอลัมน์ ก็ใช้ loc กำหนดชื่อ label **pd.loc[["row1","row2","row3"]][["col1","col2","col3"]]**
- หากต้องการเลือกบางแถวและบางคอลัมน์ ก็ใช้ iloc กำหนดตำแหน่ง position **pd.iloc[[row1,row2,row3]][["col1","col2","col3"]]**

```python
c10 = {
    "one":np.random.randint(1,30,6),
    "two":np.random.randint(1,30,6),
    "three":np.random.randint(1,30,6),
    "four":np.random.randint(1,30,6),
    "five":np.random.randint(1,30,6),
}

c11 = pd.DataFrame(c10,index=list("abcdef"))
display(c11 , c11.loc[["a","c"]][["one","two","three"]] , c11.iloc[[3,4,5]][["four","five"]])
```

#### การใช้ loc และ iloc ร่วมกับ index slicing
- หากต้องการระบุช่วงแถวและคอลัมน์แบบ label index ให้ใช้ร่วมกับ loc โดยกรณีนี้จะใช้วงเล็บ [] เพียงอันเดียว และคั่นระหว่างช่วงแถวและคอลัมน์ด้วยเครื่องหมาย , **pd.loc[start_row:stop_row:step_row,start_col:stop_col,step_col]**


- หากต้องการระบุช่วงแถวและคอลัมน์แบบ position index ให้ใช้ร่วมกับ iloc ทั้งลำดับแถวและคอลัมน์ ต้องกำหนดลำดับเป็นตัวเลขเท่านั้น **pd.iloc[start_row:stop_row:step_row,start_col:stop_col,step_col]**

```python
d1 = {
    "one":np.random.randint(1,30,6),
    "two":np.random.randint(1,30,6),
    "three":np.random.randint(1,30,6),
    "four":np.random.randint(1,30,6),
    "five":np.random.randint(1,30,6),
}

d2 = pd.DataFrame(d1,index=list("abcdef"))
display(d2 , d2.loc["a":"d":1,"one"::2] , d2.iloc[2:5,2:])
```

#### การใช้ loc และ iloc ร่วมกับ index slicing + fancy index
- กรณีใช้ loc ให้ระบุช่วงแถวเป็น label index ในแบบ index slicing และระบุชื่อคอลัมน์ที่ต้องการในแบบ fancy index , **pd.loc[start_row:stop_row:step_row,["col1","col2","col3"]]**

- กรณีใช้ iloc ให้ระบุช่วงแถวเป็น position index ในแบบ index slicing และระบุชื่อคอลัมน์ที่ต้องการในแบบ fancy index , **pd.loc[start_row:stop_row:step_row,[col1,col2,col3]]**


```python
d3 = {
    "one":np.random.randint(1,30,6),
    "two":np.random.randint(1,30,6),
    "three":np.random.randint(1,30,6),
    "four":np.random.randint(1,30,6),
    "five":np.random.randint(1,30,6),
}

d4 = pd.DataFrame(d3,index=list("abcdef"))
display(d4 , d4.loc["a":"d":1,["one","two","three"]] , d4.iloc[2:5,[0,2,4]])
```

#### การเลือกข้อมูลแบบ boolean index (masking)
- สามารถคัดกรองเอาเฉพาะข้อมูลที่ตรงเงื่อนไขได้ ด้วยโอเปอเรเตอร์ (> < != ==)

```python
d5 = {
    "one":np.array([10,30,50,70]),
    "two":np.array([20,40,60,80]),
    "three":np.array([10,40,70,100])
}

d6 = pd.DataFrame(d5,index=list("abcd"))
display(d6 , d6[d6["one"] > 40] , d6[(d6.one >= 30) & (d6.one <= 50)] , d6[["one","two"]][d6.two == 60])
```

- สามารถใช้ร่วมกับ all() และ any() ได้
- **all()** ตรวจสอบว่าข้อมูล**ทั้งหมด**ตรงเงื่อนไขหรือไม่  คืนค่า True/False
- **any(**) ตรวจสอบว่าข้อมูล**บางส่วน**ตรงเงื่อนไขหรือไม่ คืนค่า True/False

```python
d7 = {
    "one":np.array([10,30,50,70]),
    "two":np.array([20,40,60,80]),
    "three":np.array([10,40,70,100])
}

d8 = pd.DataFrame(d7,index=list("abcd"))
display(d8 , (d8.one % 2 == 0).all() , (d8["three"] >= 70).any() , (d8["three"] % 7 == 0).all())
```


### การเข้าถึงข้อมูลโดยวิธีการวนรอบ (iteration)
- หากต้องการเข้าถึงข้อมูลใน dataframe แบบวนซ้ำ เมื่อดำเนินการด้วยเหตุผลใดก็ตาม
- หากต้องการอ่านชื่อคอลัมน์ สามารถเข้าถึงโดยใช้แอตทริบิวต์ **keys**

```python
e1 = {
    "one":np.array([10,30,50,70]),
    "two":np.array([20,40,60,80]),
    "three":np.array([10,40,70,100])
}

e2 = pd.DataFrame(e1,index=list("abcd"))
for col in e1:
    print(col)
print("----------------------")
for col in e1.keys():
    print(col)
```

- ฟังชันก์ **iterrows()** จะคืนค่าของแต่ละแถวมาเป็นออบเจ็กต์ series แต่ในทางปฎิบัติอาจจะนำไปใช้งานในขั้นตอนต่อไปได้ยาก

```python
e3 = {
    "one":np.array([10,30,50,70]),
    "two":np.array([20,40,60,80]),
    "three":np.array([10,40,70,100])
}

e4 = pd.DataFrame(e3,index=list("abcd"))
display(e4)

for col in e4.iterrows():
    display(col)
```

- **itertuples()** จะคืนค่าของแต่ละแถวในแบบทูเพิล ซึ่งสามารถอ้างอิงในแต่ละคอลัมน์ได้โดยตรง

```python
e5 = {
    "one":np.array([10,30,50,70]),
    "two":np.array([20,40,60,80]),
    "three":np.array([10,40,70,100])
}

e6 = pd.DataFrame(e5,index=list("abcd"))
display(e6)

for col in e6.itertuples():
    print(col)
    print(col.Index , col.one , col.two , col.three)
```
