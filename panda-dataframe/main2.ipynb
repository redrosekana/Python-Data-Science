```python
import numpy as np
import pandas as pd
```

### ฟัง์ชันพื้นฐานของ dataFrame (ต่อ)

```python
a1 = {"one":np.random.randint(1,10,5),
      "two":np.random.randint(1,10,5),
      "three":np.random.randint(1,10,5)
}

a2 = pd.DataFrame(a1,dtype="int64")
a2
```

- **index** อ่านหรือกำหนด index
- **columns** อ่านหรือกำหนดชื่อคอลัมน์ (สามารถกำหนดช่วงคอลัมน์แบบ slicing ได้)
- **values** อ่านหรือกำหนดข้อมูล (สามารถกำหนดช่วงแถวแบบ slicing ได้)

```python
print("index เดิม = ",a2.index)
print("columns เดิม = ",a2.columns)
print("values = ",a2.values)

a2.index = list("abcde")
a2.columns = list("ABC")
print()

print("index ใหม่ = ",a2.index)
print("columns ใหม่ = ",a2.columns)
print("values = ",a2.values[1:3])
```

- **dtypes** ชนิดข้อมูลของแต่ละคอลัมน์
- **ndim** ขนาด(มิิติ) ของอาร์เรย์ที่ใช้ในการสร้าง DataFrame
- **size** จำนวนรายการข้อมูลทั้งหมดใน DataFrame
- **shape** ขนาด (จำนวนแถว , จำนวนคอลัมน์) ของ DataFrame
- **empty** ตรวจสอบว่าเป็น DataFrame ที่ว่างเปล่า (ไม่มีข้อมูล) หรือไม่ ถ้าใช่จะคืนค่า True

```python
print(a2.dtypes)
print("ndim =",a2.ndim)
print("size =",a2.size)
print("shape =",a2.shape)
print("empty =",a2.empty)
```

### ลักษณะแนวแกนของ DataFrame
- **axis = 0** ดำเนินการระหว่างแถวในแนวตั้ง

### จัดการข้อมูลที่มีค่า NaN (ฟังชันก์ที่น่าสนใจ)
   - **isnull()** ตรวจสอบว่ามีเซลล์ใดบ้างที่มีค่าเป็น NaN
   - **dropna()** ลบรายการข้อมูลที่มีค่า NaN ทิ้งไป
   - **fillna()** เปลี่ยนข้อมูลในเซลล์ที่มีค่า NaN เป็นค่าที่ระบุ (มีให้เลือกหลายรูปแบบ)

- **isnull()** ถ้าเซล์ใดเป็น NaN จะแสดงค่าเป็น True แต่ถ้าไม่เป็น NaN จะแสดงค่าเป็น False

```python
a3 = {'col1': [10, np.nan, np.nan, 20, 30],
      'col2': [np.nan, 50, np.nan, 60, np.nan],
      'col3': [np.nan, np.nan, np.nan, np.nan, np.nan]}

a4 = pd.DataFrame(a3)
display(a4 , a4.isnull())
```

- **dropna()** ควรเลือกด้วยว่าจะดำเนินการกับข้อมูลในแกนใด
    - **dropna(axis=0)** เป็นค่า default ถ้าแถวใดที่มีอย่างน้อย 1 เซลล์ที่มีค่าเป็น NaN แถวนั้นจะถูกลบทิ้ง
    - **dropna(axis=1)** ถ้าคอลัมน์ใดที่มีอย่างน้อย 1 เซลล์ที่มีค่าเป็น NaN คอลัมน์นั้นจะถูกลบทิ้ง

```python
a5 = {'col1': [10, np.nan, np.nan, 20, 30],
      'col2': [np.nan, 50, np.nan, 60, np.nan],
      'col3': [11, 22, 33, 44, 55]
     }

a6 = pd.DataFrame(a5)
display(a6 , a6.dropna(axis=0) , a6.dropna(axis=1))
```

- **fillna()** สามารถกำหนดได้ว่าจะเติมค่าอะไรลงไปแทน NaN โดยระบุอาร์กิวเมนต์ดังนี้
    - **fillna(value)** เติมค่าที่ระบุในทุกเซลล์ที่มีค่า NaN ด้วยค่าเดียวกันทั้งหมด
    - **fillna(medthod="ffill")** หรือ **fillna(medthod="pad")** เรียกว่า fill value forward เป็นการนำค่าจากเซลล์ก่อนหน้านี้ที่ไม่ใช่ NaN(ถ้ามี) มาเติมลงในเซลล์ถัดไปที่มีค่า NaN
    - **fillna(medthod="bfill")** เรียกว่า fill value backward เป็นการนำค่าจากเซลล์ถัดไปที่ไม่ใช่ NaN(ถ้ามี) ย้อนกลับมาเติมลงในเซลล์ก่อนนี้ที่มีค่า NaN
    - สามารถระบุแกนด้วย axis=0 axis=1 ได้

```python
a7 = {'col1': [10, np.nan, np.nan, 20, 30, np.nan],
      'col2': [50, np.nan, 60, np.nan, np.nan, np.nan],
      'col3': [np.nan, 70, 80, 90, np.nan, np.nan]
     }

a8 = pd.DataFrame(a7)
display(a8 , a8.fillna(0))
```

```python
display(a8 , a8.fillna(method="ffill",axis=0) , a8.fillna(method="bfill",axis=0))
```

```python
display(a8 , a8.fillna(method="ffill",axis=1) , a8.fillna(method="bfill",axis=1))
```

### ฟังชันก์เกี่ยวกับ index
- **reindex()** สำหรับการเลือกข้อมูลในแถวหรือคอลัมน์ ที่มี index ตามที่ระบุ โดยสามารถจัดเรียงลำดับตามต้องการ
- **set_index** สำหรับกำหนดให้คอลัมน์อันใดอันหนึ่งเป็น index
- **reset_index()** สำหรับต้องการกลับไปใช้ index ที่เป็นเลขลำดับ
- **rename()** สำหรับแก้ไขชื่อindex หรือ ชื่อคอลัมน์ โดยแมปชื่อที่ตรงกันในแบบดิกชินนารี
- **idxmin()** index หรือ คอลัมน์ ของข้อมูลที่มีค่าต่ำสุด
- **idxmax()** index หรือ คอลัมน์ ของข้อมูลที่มีค่าสูงสุด
- **head()** เลือกข้อมูลในช่วงเริ่มต้นตามจำนวนที่ระบุ
- **tail()** เลือกข้อมูลในช่วงท้ายตามจำนวนที่ระบุ

- **reindex()** อาจจะใช้ในการเลือกแถว หรือ เลือกคอลัมน์ก็ได้
    - หากเลือกแถว หรือ เลือกคอลัมน์ ที่ไม่มีอยู่จริงจะนำค่า NaN มาแสดง

```python
b1 = {'col1': [10, 20, 30, 40, 50],
      'col2': [60, 70, 80, 90, 100],
      'col3': [11, 22, 33, 44, 55]
     }
b2 = pd.DataFrame(b1 , index=list("abcde"))
display(b2 , b2.reindex(index=["a","d"]) , b2.reindex(columns=["col1","col2"]) , b2.reindex(index=["d","e"] , columns=["col3"]))
```

- **set_index** ใช้เมื่อต้องการเลือกคอลัมน์มาเป็น index แทนอันเดิม
    - หากกำหนดคอลัมน์ที่ไม่มีอยู่จริงจะเกิด error
- **reset_index()** ใช้เมื่อต้องการกลับไปใช้ index ที่เป็นเลขลำดับเดิม (คอลัมน์ index เดิมจะกลายเป็นคอลัมน์ข้อมูล)

```python
b3 = {'col1': [10, 20, 30, 40, 50],
      'col2': [60, 70, 80, 90, 100],
      'col3': [11, 22, 33, 44, 55]
     }
b4 = pd.DataFrame(b3 , index=list("abcde"))
display(b4 , b4.set_index("col1") , b4.reset_index())
```

- **rename()** สำหรับแก้ไขชื่อindex หรือ ชื่อคอลัมน์
    - กำหนดค่าในรูปแบบดิกชินนารี
    - ไม่จำเป็นต้องเปลี่ยนครบทุกแถว เปลี่ยนแค่บางแถวก็ได้
    - สามารถเปลี่ยนชื่อ columns ได้เช่นกัน
    - ถ้ากำหนดชื่อ index หรือ columns ที่ไม่มีจริงก็ไม่เปลี่ยนแปลง ไม่ error

```python
b5 = {'col1': [10, 20, 30, 40, 50],
      'col2': [60, 70, 80, 90, 100],
      'col3': [11, 22, 33, 44, 55]
     }
b6 = pd.DataFrame(b5 , index=list("abcde"))
display(b6 , b6.rename(index={"a":"A","b":"B"}) , b6.rename(columns={"col1":"col10"}) , b6.rename(index={"a":"A"},columns={"col1":"col10"}) )
```

- **idxmin() และ idxmax()**
    - ระบุ axis=0 จะได้ index ที่ข้อมูลมีค่าต่ำสุด/สูงสุดในแต่ละคอลัมน์
    - ระบุ axis=1 จะได้ columns ที่ข้อมูลมีค่าต่ำสุด/สูงสุดในแต่ละแถว

```python
b7 = pd.DataFrame(data=np.random.randint(10, 100, (3, 3)),
                  index=list("abc"),
                  columns=list("xyz"))

display(b7,
        b7.idxmin(axis=0),
        b7.idxmax(axis=0),
        b7.idxmin(axis=1),
        b7.idxmax(axis=1))
```

- **head(n)** n คือรายการข้อมูลเริ่มจากแถวแรก
- **tail(n)** n คือรายการข้อมูลเริ่มจากแถวสุดท้าย

```python
display(b7.head(2) , b7.tail(1))
```

### ฟังชันก์เกี่ยวกับการเปลี่ยนแปลงข้อมูล
- **insert()** เพิ่มคอลัมน์ใหม่ โดยกำหนดข้อมูลเป็นค่าเดียวกันทั้งหมด
- **assign()** เพิ่มคอลัมน์ใหม่ หรือแก้ไขข้อมูลในคอลัมน์เดิม
- **replace()** แทนที่ข้อมูลเดิมด้วยข้อมูลใหม่ในคอลัมน์ที่ระบุ
- **mask()** แทนที่ข้อมูลตามเงื่อนไขที่ระบุ
- **where()** แทนที่ข้อมูลที่ไม่ตรงเงื่อนไขที่ระบุ
- **drop()** ลบข้อมูลในคอลัมน์ หรือ แถวที่ระบุ
- **drop_duplicates()** ลบข้อมูลที่ซ้ำกัน
- **truncate()** ลบข้อมูลก่อนลำดับ หรือ หลังลำดับที่ระบุ

- **insert(ลำดับ , "ชื่อคอลัมน์" , ข้อมูล)**
    - ข้อมูลต้องครบตามจำนวน index ไม่งั้นจะ error
    - เซฟค่าในตัวแปรเดิม

```python
c1 = {'one': [10, 20, 30, 40, 50],
        'two': [60, 70, 80, 90, 100]}

c2 = pd.DataFrame(c1, index=list('abcde'))
display(c2)

c2.insert(2 , "three" , [110,120,130,140,150])
c2.insert(3 , "four" , c2.one * 5) # นำค่าจากคอลัมน์ 1 มา * 5

display(c2)
```

- **assign(ชื่อคอลัมน์=ข้อมูล)**
    - ถ้าชื่อลัมน์ซ้ำที่มีอยู่แล้วจะเป็นการแก้ไขหรือแทนที่ข้อมูล
    - ถ้าชื่อลัมน์ไม่ซ้ำที่มีอยู่แล้วจะเป็นเพิ่มคอลัมน์ในลำดับสุดท้าย
    - ไม่เซฟค่าในตัวแปรเดิม

```python
c3 = {'col1': [10, 20, 30, 40, 50],
        'col2': [60, 70, 80, 90, 100]}

c4 = pd.DataFrame(c3, index=list('abcde'))
tmp1 = c4.assign(col3=np.arange(1,6))
tmp2 = tmp1.assign(col4=np.arange(10,15))
tmp3 = tmp2.assign(col1=np.random.randint(1,100,5))

display(c4 , tmp1 , tmp2 , tmp3)
```

- **replace()** แทนที่ข้อมูลเดิมด้วยข้อมูลใหม่ในคอลัมน์ที่ระบุ มีหลายทางเลือก
    - **replace(ข้อมูลเดิม , ค่าใหม่)**
    - **replace({ข้อมูลเดิม1:ค่าใหม่1 , ข้อมูลเดิม2:ค่าใหม่2})**
    - ถ้าหากเพิ่มอาร์กิวเมนต์ **inplace=True** จะเป็นการเซฟค่าในตัวแปรเดิม แต่ถ้าไม่กำหนดต้องมีตัวแปรมารับค่า โดย defalut เป็น **inplace=False**

```python
c5 = {'col1': [10, 20, 30, np.nan, 20],
      'col2': [30, np.nan, 20, np.nan, 10]
     }

c6 = pd.DataFrame(c5, index=list('abcde'))
display(c6)

c6.replace(np.nan , 0 , inplace=True)
c6.replace({30:50} , inplace=True)

display(c6)
```

- **mask(เงื่อนไข , ค่าสำหรับแทนที่)** แทนที่ข้อมูลตามเงื่อนไขที่ระบุ เมื่อเงื่อนไขเป็น**จริง**
- **where(เงื่อนไข , ค่าสำหรับแทนที่)** แทนที่ข้อมูลที่ไม่ตรงเงื่อนไขที่ระบุ เมื่อเงื่อนไขเป็น**เท็จ**
    - ไว้เป็นทางเลือกกรณีเปรียบเทียบเงื่อนไขที่เป็นจริงมีความยุ่งยาก
    - ไม่เซฟค่าในตัวแปรเดิม
    - ถ้าหากเพิ่มอาร์กิวเมนต์ **inplace=True** จะเป็นการเซฟค่าในตัวแปรเดิม แต่ถ้าไม่กำหนดต้องมีตัวแปรมารับค่า โดย defalut เป็น **inplace=False**

```python
c7 = {'col1': [10, -20, 30, -40, 50],
      'col2': [11, -22, -33, 44, -55],
      'col3': [-60, -70, 80, 90, -100]
     }
c8 = pd.DataFrame(c7, index=list('abcde'))
display(c8)

tmp4 = c8.mask(c8 < 0 , 100)
tmp5 = tmp4.where(tmp4 > 10 , 2000)

display(tmp4 , tmp5)
```

- **drop(index=[] , columns=[])** ลบข้อมูลในคอลัมน์ หรือ แถวที่ระบุ อาจจะระบุแค่ index หรือ columns อย่างใดอย่างหนึ่งก็ได้
    - ไม่เซฟค่าในตัวแปรเดิม
    - ถ้าหากเพิ่มอาร์กิวเมนต์ **inplace=True** จะเป็นการเซฟค่าในตัวแปรเดิม แต่ถ้าไม่กำหนดต้องมีตัวแปรมารับค่า โดย defalut เป็น **inplace=False**
- **drop_duplicates("คอลัมน์")** ให้ระบุชื่อคอลัมน์ซึ่งต้องการลบข้อมูลที่ซ้ำกันออกไป
- **truncate(before,after)** ลบข้อมูลก่อนลำดับ หรือ หลังลำดับที่ระบุ
    - ให้ลบแถวก่อน before และ แถวหลังลำดับ after ซึ่งอาจจะมีเพียงอย่างใดอย่างหนึ่งก็ได้

```python
c9 = {'col1': [10, 20, 30, 40, 50],
      'col2': [60, 70, 80, 90, 100],
      'col3': [11, 22, 33, 44, 55]
     }

c10 = pd.DataFrame(c9, index=list('abcde'))
display(c10)

tmp6 = c10.drop(index=["a","b"],columns=["col2"])

display(tmp6)
```

```python
c11 = {'col1': np.random.choice([10,20,30],5),
       'col2': np.random.choice([10,20,30],5)
      }

c12 = pd.DataFrame(c11, index=list('abcde'))
display(c12)

tmp7 = c12.drop_duplicates("col1")

display(tmp7)
```

```python
c13 = pd.DataFrame(data=np.random.randint(10, 100, (7, 3)),
                  index=list('abcdefg'),
                  columns=['one', 'two', 'three'])

display(c13)
c13.truncate(before='c', after='e')
```
